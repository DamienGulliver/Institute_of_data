# Lab 4.1.2 Linear Regression

## Prediction of House Price Using Linear Regression

### Data

The Sacramento real estate transactions file is a list of 985 real estate transactions in the Sacramento area reported over a five-day period, as reported by the Sacramento Bee.

````python
import pandas as pd
import numpy as np
import pandas as pd
from sklearn import metrics


%matplotlib inline
````

### 1. Explore Data

Perform EDA on this dataset. Find Dependent and Independent Variables to predict house price.

````python
# Read CSV
house_csv = pd.read_csv("/Users/damien/Documents/institute_of_data/dat/mod4/Sacramento_transactions.csv")
````

````python
house_csv.head()
````

````python
house_csv.tail()
````

````python
house_csv.info()
````

#### Dropping Null Values

````python
house_csv['beds'].replace([0], np.NaN, inplace=True)
house_csv['baths'].replace([0], np.NaN, inplace=True)
house_csv['sq__ft'].replace([0], np.NaN, inplace=True)
````

````python
house_csv.isnull().sum()
````

````python
house_csv = house_csv.dropna()
````

````python
house_csv.shape
````

### 2. Predict Price

We are going to predict `price` from available information.

#### 2.1 What is the Target Variable? Why?

### ANSWER

Price is the target variable because we are trying to predict it from other independent variables

#### 2.2 List all possible variables which might be Independent/Predictor variable.

````python
# ANSWER

#All features apart from price could be predictor variables

house_csv.columns
````

#### 2.3 Find correlation between variables

Find correlation between variables. Which is the best predictor? Why? State your reason.

````python
# ANSWER

#sq_ft is the best predictor because it has the highest correlation coefficient

house_csv.corr()['price'].sort_values()
````

#### 2.4 Find Coefficient and Intercept using Linear Regression

Using LinearRegression of sklearn.linear_model package find Coefficient and Intercept.

Create an instance of LinearRegression.

Explore following methods:

- fit
- predict
- score

````python
from sklearn.linear_model import LinearRegression
````

##### 2.4.1 Fit predictor and target variables using linear regression 

````python
# ANSWER

X = house_csv['sq__ft'].values.reshape(-1,1)
Y = house_csv['price']

reg = LinearRegression()
reg.fit(X, Y)
reg.predict(X)
````

##### 2.4.2 Find R Score

Find $R^2$ Using ```score``` method of Linear Regression.

````python
# ANSWER
reg.score(X, Y)
````

### 3. Splitting Data

````python
from sklearn.model_selection import train_test_split
````

#### 3.1 Create training and testing subsets

Hint: Using `train_test_split` Library.

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

````python
# ANSWER
## Create training and testing subsets
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
````

#### 3.2 Check Shape, Sample of Test Train Data

````python
# ANSWER
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
````

#### 3.3 Using Linear Regression Find The Score

1. Fit model using X_train, y_train
2. Find score using X_test, y_test

````python
# ANSWER
reg.fit(X_train, y_train)
print('intercept:', reg.intercept_)
print('r2:', reg.coef_)
````

````python
# ANSWER
y_pred = reg.predict(X_test)
reg.score(X_test, y_test)
````

### 4. Predict House Price

Let's assume we have information of following house:

- street:	1140 EDMONTON DR
- city:	SACRAMENTO
- zip:	95833
- state:	CA
- beds:	3
- baths:	2
- sq__ft:	1204
- type:	Residential

**Predict the price of this house using linear regression model.**

````python
# ANSWER
reg.predict([[1204]])
````

#### Find the error

````python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
````
